```python
import matplotlib.pyplot as plt
import networkx as nx

class NewGraph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, node1, node2):
        if node1 in self.graph:
            self.graph[node1].append(node2)
        else:
            self.graph[node1] = [node2]

    def dfs_traversal(self, start):
        visited = set()

        def dfs_recursive(node):
            if node not in visited:
                print(node, end=' ')
                visited.add(node)
                if node in self.graph:
                    for neighbor in self.graph[node]:
                        if neighbor not in visited:
                            dfs_recursive(neighbor)

        dfs_recursive(start)

    def draw_graph(self):
        G = nx.DiGraph()

        for node, neighbors in self.graph.items():
            G.add_node(node)
            for neighbor in neighbors:
                G.add_edge(node, neighbor)

        pos = nx.spring_layout(G)
        labels = {node: str(node) for node in G.nodes()}

        nx.draw(G, pos, with_labels=True, labels=labels, node_size=500, node_color='lightgreen', font_size=10)
        plt.title("Grafo Nuevo")
        plt.show()

# Ejemplo de uso
new_graph = NewGraph()
new_graph.add_edge('A', 'B')
new_graph.add_edge('A', 'C')
new_graph.add_edge('B', 'C')
new_graph.add_edge('C', 'A')
new_graph.add_edge('C', 'D')
new_graph.add_edge('D', 'D')

print("Recorrido DFS a partir del nodo 'C':")
new_graph.dfs_traversal('C')

# Dibujar el grafo
new_graph.draw_graph()
```
